```python
import statsmodels.formula.api as smf
model = smf.ols('y ~ x2', data=data)
res = model.fit()
print(res.summary())
```

По полученной модели коэффициент детерминации (R- squared) показывает, что 30,1% вариации переменной у зависит от фактора х2. Оставшиеся 69,9% (100-R2) приходится на неучтенные факторы.

Уравнение регрессии имеет вид: y=a+bx

В рассмотренном примере: y=29.9068+2.2109x

Значимость уравнения в целом проводится по критерию F-Фишера. Для этого значение Prob(F-statistic) сравниваем значения с принимаемым уровнем значимости(вероятность ошибки) - 5%.
Так как Prob(F-statistic) имеет значение 0.0000795, что ниже, чем 0,05 - делается вывод, что уравнение статистически значимо и может применяться в экономических исследованиях.

Значимость параметров проверяем по критерию t-Стьюдента. 
Так как значения в столбце P>|t| ниже 0,05 - делается вывод, что параметр значим и имеет экономический смысл.

Так как Prob(F-statistic) имеет значение 0.00002222, что ниже, чем 0,05 - делается вывод, что уравнение статистически значимо и может применяться в экономических исследованиях.

Значимость параметров проверяем по критерию t-Стьюдента. 
Так как значения в столбце P>|t| ниже 0,05 - делается вывод, что параметр значим и имеет экономический смысл. 

```python
import numpy as np
np.cov(data['x2'],data['y'])
```

```python
def cov(data,k,t):
    xsr = data[k].mean()
    ysr = data[t].mean()
    sum = 0
    leng = len(data[k])
    for i in range(leng):
        sum += (data[k][i] - xsr)*(data[t][i] - ysr)
    return(sum/(leng-1))
```

```python
cov(data,'x2','y')
```

```python
np.corrcoef(data['x2'],data['y'])
```

```python
def corr(data,k,t):
    xsr = data[k].mean()
    ysr = data[t].mean()
    leng = len(data[k])
    sum1 = 0
    sum2 = 0
    for i in range(leng):
        sum1 += (data[k][i]-xsr)**2
        sum2 += (data[t][i]-ysr)**2
    
    sx = (sum1/(leng-1))**0.5
    sy = (sum2/(leng-1))**0.5
    
    r = cov(data,k,t)/(sx*sy)
    return(r)
```
