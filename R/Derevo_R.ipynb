```r
library(rpart)
library(rpart.plot)
```

```r
dt <- read.csv('heart.csv', header = T)
```

```r
dt
```

```r
dt_mixed <- dt[order(runif(232)),]
```

```r
dt_mixed
```

```r
train <- dt_mixed[1:150,]
```

```r
test <- dt_mixed[151:232,]
```

```r
test
```

```r
tree <- rpart(train$target~.,data=train,method='class')
```

```r
rpart.plot(tree)
```

```r
pred <- predict(tree,test,type='class')
```

```r
pred
```

```r
table(pred,test$target)
```

```r
acc<-mean(pred==test$target)
```

```r
paste(round(100*acc,2),'%')
```
